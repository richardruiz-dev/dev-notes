## File Management

```python
some_file = open("test.txt","w+")
some_file.write("\tline 1:This is the first test string\n\t\tline 2\n\t\tline 3")
some_file.flush()
some_file.seek(0)
print(some_file)

print("read 1:", some_file.read())
print("read 2:", some_file.read()) #<-- will not print anything

some_file_again= open("test.txt")
print("read 3:",some_file_again.readline(7))
print("read 4:",some_file_again.readline())
some_file_again.seek(0)
print("read 5:",some_file_again.readline())

some_file_again.close()
some_file_again = open("test.txt", 'w+')
some_file_again.write("This is the second test string.\nline2\nline3")
some_file_again.flush()
some_file_again.seek(0)
print("read 6:",some_file_again.readlines(0))
some_file_again.seek(0)
print("read 7:",some_file_again.readlines(1))
print("read 8:",some_file_again.readlines(2))
print("read 9:",some_file_again.readlines(3))
print("read 10:",some_file_again.readlines(0))
```
